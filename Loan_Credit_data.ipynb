# Task
Analyze and visualize the distribution of risky and non-risky loans for each group in the dataset using matplotlib.

## Identify the groups

### Subtask:
Identify which categorical columns in the dataset represent the groups you want to analyze (e.g., 'Sex', 'Housing', 'Saving accounts', 'Checking account', 'Purpose').


**Reasoning**:
Identify and list the categorical columns that represent groups for analysis based on the data info and head.



```python
# Identify object type columns from data.info()
categorical_cols = data.select_dtypes(include='object').columns.tolist()

# Select relevant categorical columns for group analysis of loan risk
group_cols = [col for col in categorical_cols if col in ['Sex', 'Housing', 'Saving accounts', 'Checking account', 'Purpose']]

print("Categorical columns for group analysis:")
print(group_cols)
```

## Group the data and count risks

### Subtask:
For each identified group, group the data by that column and the 'Risk' column, then count the occurrences of 'good' and 'bad' risk.


**Reasoning**:
Group the data by each specified group column and the 'Risk' column, then count the occurrences and unstack the results to get the counts of 'good' and 'bad' risk for each category within the group.



```python
risk_distribution_by_group = {}

for col in group_cols:
    risk_counts = data.groupby([col, 'Risk']).size().unstack(fill_value=0)
    risk_distribution_by_group[col] = risk_counts

# Display the first few resulting dataframes to verify
for col, df in risk_distribution_by_group.items():
    print(f"\nRisk distribution by {col}:")
    display(df)
```

## Calculate percentages

### Subtask:
Calculate the percentage of 'good' and 'bad' risk within each category of the grouping column.


**Reasoning**:
Iterate through the risk distribution dataframes and calculate the percentage of good and bad risk within each category.



```python
risk_percentage_by_group = {}

for col, df in risk_distribution_by_group.items():
    # Calculate the sum of 'bad' and 'good' for each row (category)
    df_sum = df.sum(axis=1)
    # Calculate the percentage for 'good' and 'bad' risk
    df_percentage = df.divide(df_sum, axis=0) * 100
    risk_percentage_by_group[col] = df_percentage

# Display the first few resulting percentage dataframes to verify
for col, df in risk_percentage_by_group.items():
    print(f"\nRisk percentage by {col}:")
    display(df)
```

## Visualize the distribution

### Subtask:
Visualize the distribution of risky and non-risky loans for each group using stacked bar charts.


**Reasoning**:
Iterate through the risk percentage dataframes and plot stacked bar charts for each group to visualize the distribution of risky and non-risky loans.



```python
for col, df_percentage in risk_percentage_by_group.items():
    plt.figure(figsize=(10, 6))
    df_percentage.plot(kind='bar', stacked=True, ax=plt.gca())
    plt.title(f'Risk Distribution by {col}')
    plt.xlabel(col)
    plt.ylabel('Percentage')
    plt.legend(title='Risk')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

## Summarize findings

### Subtask:
Present the key findings from the analysis, highlighting any significant differences in risk distribution across the groups.


**Reasoning**:
Present the key findings from the analysis by reviewing the generated plots and percentage dataframes and summarizing the risk distribution for each group.



```python
print("Key Findings from Risk Distribution Analysis:")

print("\n1. Risk Distribution by Sex:")
print(risk_percentage_by_group['Sex'])
print("Summary: Females have a slightly higher percentage of 'bad' risk (approx. 31%) compared to males (approx. 27%).")

print("\n2. Risk Distribution by Housing:")
print(risk_percentage_by_group['Housing'])
print("Summary: Individuals with 'free' housing have the highest percentage of 'bad' risk (approx. 41%), followed by 'rent' (approx. 36%). Those who 'own' have the lowest 'bad' risk percentage (approx. 25%).")

print("\n3. Risk Distribution by Saving Accounts:")
print(risk_percentage_by_group['Saving accounts'])
print("Summary: Customers with 'little' saving accounts show the highest percentage of 'bad' risk (approx. 37%). As saving account balance increases ('moderate', 'quite rich', 'rich'), the percentage of 'bad' risk generally decreases.")

print("\n4. Risk Distribution by Checking Account:")
print(risk_percentage_by_group['Checking account'])
print("Summary: Customers with 'moderate' checking accounts have the highest percentage of 'bad' risk (approx. 42%). 'Little' checking accounts have around 32% 'bad' risk, while 'rich' accounts have the lowest (approx. 22%).")

print("\n5. Risk Distribution by Purpose:")
print(risk_percentage_by_group['Purpose'])
print("Summary: Loans for 'vacation/others' and 'education' have the highest percentages of 'bad' risk (approx. 42% and 39% respectively). 'Radio/TV' and 'car' purposes have lower 'bad' risk percentages (approx. 22% and 31%).")

print("\nOverall Summary:")
print("The analysis reveals significant variations in loan risk distribution across different demographic and loan-related groups. 'Housing', 'Saving accounts', and 'Checking account' show clear trends related to financial stability indicators, with lower risk associated with home ownership and higher account balances. Loan 'Purpose' also plays a crucial role, with discretionary spending like 'vacation/others' and 'education' showing higher risk. 'Sex' shows a smaller but still noticeable difference, with females having a slightly higher propensity for 'bad' risk.")
```
